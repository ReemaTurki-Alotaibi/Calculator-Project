```python
# calculator_with_tests_colab.py

import math

# ========================
# Calculator Functions
# ========================

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Invalid value for denominator, can't divide by 0!"
    return a / b

def square(a):
    return a * a

def power(a, b):
    return a ** b

def sqrt(a):
    if a < 0:
        return "Invalid value, can't take square root of negative number!"
    return math.sqrt(a)

# ========================
# Tests
# ========================

def test_add():
    assert add(2, 3) == 5, "Test failed: 2 + 3 should be 5"
    assert add(-1, 5) == 4, "Test failed: -1 + 5 should be 4"
    assert add(0, 0) == 0, "Test failed: 0 + 0 should be 0"
    print("All add tests passed!")

def test_subtract():
    assert subtract(5, 3) == 2, "Test failed: 5 - 3 should be 2"
    assert subtract(3, 5) == -2, "Test failed: 3 - 5 should be -2"
    assert subtract(0, 0) == 0, "Test failed: 0 - 0 should be 0"
    print("All subtract tests passed!")

def test_multiply():
    assert multiply(2, 3) == 6, "Test failed: 2 * 3 should be 6"
    assert multiply(-1, 5) == -5, "Test failed: -1 * 5 should be -5"
    assert multiply(0, 100) == 0, "Test failed: 0 * 100 should be 0"
    print("All multiply tests passed!")

def test_divide():
    assert divide(10, 2) == 5, "Test failed: 10 / 2 should be 5"
    assert divide(5, 2) == 2.5, "Test failed: 5 / 2 should be 2.5"
    assert divide(3, 0) == "Invalid value for denominator, can't divide by 0!", "Test failed: Dividing by 0 should return a warning"
    print("All divide tests passed!")

def test_square():
    assert square(4) == 16, "Test failed: 4^2 should be 16"
    assert square(-3) == 9, "Test failed: (-3)^2 should be 9"
    assert square(0) == 0, "Test failed: 0^2 should be 0"
    print("All square tests passed!")

def test_power():
    assert power(2, 3) == 8, "Test failed: 2^3 should be 8"
    assert power(5, 0) == 1, "Test failed: 5^0 should be 1"
    assert power(2, -1) == 0.5, "Test failed: 2^-1 should be 0.5"
    print("All power tests passed!")

def test_sqrt():
    assert sqrt(16) == 4, "Test failed: sqrt(16) should be 4"
    assert sqrt(0) == 0, "Test failed: sqrt(0) should be 0"
    assert sqrt(-9) == "Invalid value, can't take square root of negative number!", "Test failed: sqrt(-9) should return error message"
    print("All sqrt tests passed!")

# ========================
# Running the tests
# ========================

def run_tests():
    test_add()
    test_subtract()
    test_multiply()
    test_divide()
    test_square()
    test_power()
    test_sqrt()

# Run the tests
run_tests()
```
